```python
import requests
import json
from pprint import pprint
# Dependencies
from bs4 import BeautifulSoup
import requests
```

```python
# URL of page to be scraped
url_1 = 'https://www.imdb.com/chart/top?ref_=nv_mv_250'
```

```python
# Retrieve page with the requests module
response = requests.get(url_1)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# results are returned as an iterable list
results = soup.find_all("tr")
print(results)
```

```python
# Loop through returned results
movies=[]
for result in results:
    # Error handling
    try:
        
        # Identify and return title of listing
#         title = result.span.text
        title = result.find('td', class_='titleColumn').a.text
#         title = result.find('td', class_='titleColumn').text
        # Identify and return price of listing
#         rating = result.find('td', class_= "ratingColumn imdbRating").text
#         # Identify and return link to listing
#         year = result.tr.td.find('span').text
        
        # Print results only if title, price, and link are available
#         if (title and price and link):
        if (title):
#             print('-------------')
            movies.append(title)
            print(title)
#             print(year)
    except AttributeError as e:
        print(e)
```

```python
#base url
base_url = 'http://www.omdbapi.com/?apikey=trilogy&t='
```

```python
movies
```

```python
# movies = ['Avengers','Pulp+Fiction','Forrest+Gump']
response =[]
for movie in movies:
    url = base_url + movie
    url=url.replace(" ","+")
    response.append(requests.get(url))
    print(url)


# title = 'Avengers'
```

```python
# Perform a get request for this character
data = []
for resp in response:
    data.append(resp.json())
```

```python
# Collecting the name of the character collected
# movie_details = data["Title"]

for name in data:
    try:
        print(name['Title'],name['Genre'],name['BoxOffice'])
    except KeyError as e:
        print(e)
        
```
